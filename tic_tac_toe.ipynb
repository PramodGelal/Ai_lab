```python
def is_open_line(line, player):
    """
    Check if the given line (row, column, or diagonal) is still open for the player.
    A line is open if it contains only the player's marks or empty cells.
    """
    return all(cell == player or cell == ' ' for cell in line)

def calculate_heuristic(board):
    player = 'X'
    opponent = 'O'

    player_open_lines = 0
    opponent_open_lines = 0

    # Check rows
    for row in board:
        if is_open_line(row, player):
            player_open_lines += 1
        if is_open_line(row, opponent):
            opponent_open_lines += 1

    # Check columns
    for col in range(3):
        column = [board[row][col] for row in range(3)]
        if is_open_line(column, player):
            player_open_lines += 1
        if is_open_line(column, opponent):
            opponent_open_lines += 1

    # Check diagonals
    main_diag = [board[i][i] for i in range(3)]
    anti_diag = [board[i][2-i] for i in range(3)]

    if is_open_line(main_diag, player):
        player_open_lines += 1
    if is_open_line(main_diag, opponent):
        opponent_open_lines += 1

    if is_open_line(anti_diag, player):
        player_open_lines += 1
    if is_open_line(anti_diag, opponent):
        opponent_open_lines += 1

    # Calculate the heuristic value
    heuristic = player_open_lines - opponent_open_lines
    return heuristic

# Example of usage:

board = [
    ['X', 'O', ' '],
    ['X', ' ', ' '],
    ['O', 'X', 'O']
]

heuristic_value = calculate_heuristic(board)
print("Heuristic value of the board:", heuristic_value)
```
