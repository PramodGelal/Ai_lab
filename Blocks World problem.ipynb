```python
# Define the goal state for the Blocks World problem
goal_state = {
    'A': [0],  # A is on the table (position 0)
    'B': [1],  # B is on top of A
    'C': [2],  # C is on top of B
}

# Function to get the correct position of a block
def get_correct_position(block):
    # Return the goal position of the block
    return goal_state.get(block, None)

# Function to calculate the heuristic value for a given state
def calculate_heuristic(state):
    heuristic = 0
    visited_blocks = set()

    # Iterate over each stack in the state
    for stack in state.values():
        for i in range(len(stack)):
            block = stack[i]
            # For each block, check its support structure (the block below it)
            correct_position = get_correct_position(block)

            # The support structure for the block should match its goal position
            if i == 0:  # If the block is at the bottom of a stack (supported by the table)
                if correct_position == [i]:
                    heuristic += 1
            else:  # If the block is supported by another block
                support_block = stack[i-1]
                if correct_position == [i]:
                    heuristic += 1
                else:
                    heuristic -= 1
                visited_blocks.add(support_block)

    return heuristic


# Function to represent the current state in the Blocks World
def print_state(state):
    for stack in state.values():
        print(stack)
    print()


# Example of an initial state for the Blocks World problem
state = {
    0: ['A'],
    1: ['B'],
    2: ['C']
}

# Print initial state
print("Initial State:")
print_state(state)

# Calculate and print the heuristic for the given state
heuristic_value = calculate_heuristic(state)
print(f"Heuristic value for the current state: {heuristic_value}")
```
